```python
import pandas as pd
import string
import nltk
```

```python
df= pd.read_csv(r"C:\Users\abhim\OneDrive\Desktop\AI_Human.csv")
df.head(10)
```

```python
df['text'][0]
```

```python
df['text']= df['text'].str.lower()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['text'][0]
```

```python
def remov_punc(st):
    new= [x for x in st if x not in string.punctuation]
    new= ''.join(new)
    return new
df['text']= df['text'].apply(remov_punc)
df['text'][0]
```

```python
from nltk.corpus import stopwords
st_words= set(stopwords.words('english'))
def rem_stopwords(st):
    words= st.split()
    stwords= [x for x in words if x not in st_words]
    stwords=' '.join(stwords)
    return stwords
df['text']= df['text'].apply(rem_stopwords)
df['text'][0]
```

```python
df['text'][0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(df['text'],df['generated'],test_size=0.4,random_state=56,stratify= df['generated'])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
```

```python
pipeline= Pipeline([('cnt_vect', CountVectorizer()),('tfid',TfidfTransformer()),('naive_bayes', MultinomialNB())])
```

```python
pipeline.fit(x_train,y_train)
```

```python
y_predicted=pipeline.predict(x_test)
```

```python
print(classification_report(y_test,y_predicted))
```
